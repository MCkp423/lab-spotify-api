```python
#Once the previous function has been created, create another function add_audio_features(df, audio_features_df) to concat a given dataframe with the audio features dataframe and return the extended data frame.
```

```python
import numpy as np
import time

def get_audio_features(list_of_song_ids):
    sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="YOUR_CLIENT_ID",
                                                               client_secret="YOUR_CLIENT_SECRET"))
    audio_features = []

    chunks = np.array_split(list_of_song_ids, len(list_of_song_ids) // 50 + 1)

    for chunk in chunks:
        print("Collecting IDs for chunk...")
        features = sp.audio_features(chunk)
        audio_features.extend(features)
        time.sleep(20)  

    return audio_features
```

```python
list_of_song_ids = hot_songs_df['id'].tolist()
audio_features = get_audio_features(list_of_song_ids)

audio_features_df = pd.DataFrame(audio_features)
```

```python
list_of_not_hot_song_ids = not_so_hot_songs_df['id'].tolist()
not_hot_audio_features = get_audio_features(list_of_not_hot_song_ids)

not_hot_audio_features_df = pd.DataFrame(not_hot_audio_features)
```

```python
#Finally, replace the old internal files of songs (hot and not hot) with the extended data frames with the audio features and save them into separate files on the disk.
```

```python
def add_audio_features(songs_df, audio_features_df):
    return pd.concat([songs_df, audio_features_df], axis=1)
```

```python
extended_hot_songs_df = add_audio_features(hot_songs_df, audio_features_df)
extended_not_so_hot_songs_df = add_audio_features(not_so_hot_songs_df, not_hot_audio_features_df)

extended_hot_songs_df.to_csv('hot_songs_with_audio_features.csv', index=False)
extended_not_so_hot_songs_df.to_csv('not_so_hot_songs_with_audio_features.csv', index=False)
```
